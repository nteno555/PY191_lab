##### Trimmed Average Function

```python
import math

def trimmed_average(ls):
    if (len(ls) < 2): return 0
    ls.sort()
    sum = 0
    for i in range(1, len(ls)-1):
        sum += ls[i]
    return sum/(len(ls)-2)

ls = [2, 4, 1, 3, 7, 1]

print(trimmed_average(ls))
```

##### Find Factor Function

```python
def find_factors(n):
    factors = []
    square_root = int(math.sqrt(n)) + 1
    for i in range(1, square_root):
        if (n % i == 0):
            factors.append(i)
            if (n//i != i): factors.append(n//i)
    factors.sort()
    return factors

print(find_factors(25))
```
