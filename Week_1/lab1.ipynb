## 3. Variables & Data Types
Python is **dynamically typed** – variable types are inferred at runtime.

| Type | Example | Literal |
|------|---------|---------|
| `int` | whole numbers | `x = 42` |
| `float` | decimal numbers | `pi = 3.14159` |
| `str` | text | `name = "Ada"` |
| `bool` | truth values | `is_valid = True` |
| `list` | ordered, mutable collection | `scores = [95, 88, 72]` |
| `dict` | key‑value pairs | `user = {'id': 1, 'name': 'Sam'}` |

Run the cell below to see the types in action.

```python
x = 42
pi = 3.14159
name = "Ada"
is_valid = True
scores = [95, 88, 72]
user = {'id': 1, 'name': 'Sam'}

print(type(x), type(pi), type(name), type(is_valid), type(scores), type(user))
```

```python
# 1. List Comprehension Speed Up
import timeit

# For Loop Approach
def for_loop():
    nums = list(range(1, 100001))
    for i in range(0, 100000):
        nums[i] = nums[i] ** 2
    return nums

def list_comprehension():
    nums = list(range(1, 100001))
    squares = [val ** 2 for val in nums]

for_loop_time = timeit.timeit(for_loop, number=1)
list_comprehension_time = timeit.timeit(list_comprehension, number=1)

print(f"For Loop Time: {for_loop_time}")
print(f"List Comprehension Time: {list_comprehension_time}")
```

```python
# 2. Dictionary Comprehension Puzzle
alphabet = "abcdefghijklmnopqrstuvwxyz"
dictionary = {}
for i in range(0, len(alphabet)):
    dictionary[alphabet[i]] = i

print(dictionary)
```

```python
# 3. NumPy Broadcasting Art
import numpy as np
import matplotlib.pyplot as plt

n = 5
array = np.zeros((n, n))
midpoint = (n+1)//2 - 1

array[midpoint][midpoint] = 0
for i in range(0, n):
    for j in range(0, n):
        if (i == midpoint and j==midpoint): continue
        array[i][j] = abs(i-midpoint) + abs(j-midpoint)

plt.imshow(array)
```

```python
# 4. Prime Checker Function
import math

def is_prime(n):
    sqrt = int(math.sqrt(n))
    for i in range(2, sqrt+1):
        if (n%i == 0): return False
    return True

print(is_prime(2))
print(is_prime(3))
print(is_prime(7))
print(is_prime(9973))
```
