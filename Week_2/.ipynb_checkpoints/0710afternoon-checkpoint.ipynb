```python
import pandas as pd

stock = pd.read_csv("RBLX.csv")
stock = stock.drop(0)

close = stock["Close"].to_numpy()
sma = stock["Short_MA"].to_numpy()
lma = stock["Long_MA"].to_numpy()

balance = 100
max_bal = 100
stocks = 0
trades = 0

for i in range(0, len(close)):
    if (sma[i] > lma[i]):
        stocks += balance/float(close[i])
        balance = 0
        trades += 1
    elif (sma[i] < lma[i]):
        balance += stocks * float(close[i])
        max_bal = max(balance, max_bal)
        stocks = 0
        trades += 1

print(max_bal)
print(trades)
```
